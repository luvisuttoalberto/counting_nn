```python
%load_ext autoreload
%autoreload 2
from utilities import generator
import matplotlib.pyplot as plt
```

```python
n_images = 10
r_rows = 2
images, counts = generator(n_images=n_images, method='classic')
fig, ax = plt.subplots(nrows=r_rows, ncols=int(n_images/r_rows), figsize=(16, int(16/(n_images/(r_rows*2)))))
for index_row, row in enumerate(ax):
    for index_col, col in enumerate(row):
        col.imshow(images[index_row + index_col])
        col.set_title('n = ' + str(counts[index_row + index_col] + 1))
        col.axis('off')
```
